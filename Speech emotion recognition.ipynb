# Load the Dataset

```python
paths=[]
labels=[]
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        paths.append(os.path.join(dirname, filename))
        label = filename.split('_')[-1]
        label=label.split('.')[0]
        labels.append(label.lower())
print('Dataset is loaded')
        
```
